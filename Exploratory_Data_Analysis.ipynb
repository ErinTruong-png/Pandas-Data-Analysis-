Question: Are there certain continents that grow faster in population?

```python
#Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import data
df = pd.read_csv('/content/world_population.csv')
print(df.columns)
```

```python
df
```

```python
sns.heatmap(df.corr(numeric_only = True), annot = True)
plt.rcParams['figure.figsize'] = (20, 7)
plt.show()
```

```python
#Group by continents to see which continents have done well in population
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population',]].mean(numeric_only = True).sort_values(by = '2022 Population', ascending = False)
df2

#Tells us that Asia has the highest population
```

```python
#Tranpose the data to make sure that the data is transposed
df3 = df2.transpose()
df3
```
